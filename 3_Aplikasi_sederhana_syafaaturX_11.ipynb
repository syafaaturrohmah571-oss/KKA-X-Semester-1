#Aplikasi Sederhana menghitung Volume Bola

```python
jari_jari = int(input("Masukkan jari_jari dari bola(cm): "))
pi = 22/7
volume_bola = 4/3 * pi * jari_jari**3
print(f" Volume bola adalah {volume_bola} cm^3. ")
```

#Menghitung volume tabung

```python

jari_jari = float(input("Masukan jari-jari alas tabung (cm): "))
tinggi = float(input("Masukan tinggi tabung (cm): "))
pi = 22/7
volume_tabung = pi * jari_jari**2 * tinggi
print(f"volume tabung adalah {volume_tabung} cm^3. ")
```
